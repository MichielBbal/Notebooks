## 1. My first BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
result = requests.get("https://www.google.com/")
                      
print(result.status_code)
print(result.headers)
```

```python
src = result.content
print(src)
```

```python
soup = BeautifulSoup(src, 'lxml')
links = soup.find_all("a")
print(links)
```

```python
for link in links:
    if "about" in link.text:
        print(link)
        #print(link.attrs['href']) # print what it goes trough
```

```python
import requests
from bs4 import BeautifulSoup

result = requests.get("https://www.whitehouse.gov/briefings-statements")

src = result.content
print(src)
```

```python
soup = BeautifulSoup(src, 'lxml')
urls = []
for h2_tag in soup.find_all('h2'):
    a_tag = h2_tag.find('a')
    urls.append(a_tag.attrs['href'])
print(urls)
```

```python
import PyPDF2
# pdf file object
# you can find find the pdf file with complete code in below
pdfFileObj = open('COVID-19_WebSite_rapport_20200512_1143 (1).pdf', 'rb') # insert your file here
# pdf reader object
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
# number of pages in pdf
print('paginaas = '+ str(pdfReader.numPages))
# a page object
pageObj = pdfReader.getPage(9)
# extracting text from page.
# this will print the text you can also save that into String
print(pageObj.extractText())
```
