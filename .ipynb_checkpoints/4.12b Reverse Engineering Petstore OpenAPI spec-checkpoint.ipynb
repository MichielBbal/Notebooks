## 1. Getting the API and converting it to a python dict

```python
import wget

url = 'https://petstore3.swagger.io/api/v3/openapi.json'
my_oas = wget.download(url)
print(my_oas)
```

```python
#loading the API and converting it to a python dict
import json

with open(my_oas) as f:
    spec = json.load(f)
list(spec)
```

```python
# Optional: printing the entire API spec
#print(spec)
```

```python
#Getting the metadata
API_version =spec['info']['version']
API_version = "v"+API_version[:5]
OAS_version =spec['openapi']
API_name = spec['info']['title']
API_name = API_name.replace(" ", "_")
print("OAS:", OAS_version, API_name, API_version)
```

## 2. Getting the paths and endpoints

```python
# getting the paths in the 
paths =list(spec['paths'])
print(len(paths))
paths
```

```python
# combining http verbs with endpoints
endpoints_dict = {}
for i in range (len(paths)):
    endpoints_dict[paths[i]]=list(spec['paths'][paths[i]])
endpoints_dict
```

```python
from prettytable import PrettyTable
    
my_table = PrettyTable()

my_table.field_names = ["Path", "Verb", "Description"]
endpoints = list(spec['paths'])
for i in range (len(endpoints)):
    verbs = list(spec['paths'][endpoints[i]])
    for j in range (len(verbs)):
        description_string = spec['paths'][endpoints[i]][verbs[j]]['description']
        description_string = description_string.strip('\n')
        #print(verbs[j], endpoints[i], description_string)
        my_table.add_row([endpoints[i], verbs[j], description_string[:80]])
print(my_table)
```

## 3. Getting the objects

```python
spec['components']['schemas']
```

```python
#Getting the objects
objects = list(spec['components']['schemas'])
print(len(objects))
objects
```

```python
#creating a dictionary with objects and their properties
object_dict ={}
for i in range (len(objects)):
    try:
        object_dict[objects[i]]=list(spec['components']['schemas'][objects[i]]['properties'])#['properties']
    except:
        my_list= []
        my_dict = spec['components']['schemas'][objects[i]]['allOf'][0]
        for key, value in my_dict.items():
            my_string = str(value)
        my_string = my_string.replace('#/components/schemas/', 'allOf: ')
        my_list.append(my_string)
        object_dict[objects[i]] = my_list+list(spec['components']['schemas'][objects[i]]['allOf'][1]['properties'])
object_dict
```

```python
#converting the python dict to a JSON 
import json 
      
# Serializing json  
json_object = json.dumps(object_dict, indent = 4) 
print(json_object)
```

## 4. Writing it to a .json, plantuml and .csv file

```python
#writing it to a json
import json
with open(API_name+'_data.json', 'w', encoding='utf-8') as f:
    json.dump(json_object, f, ensure_ascii=False, indent=4)
```

```python
import glob
my_jsons = glob.glob('*.json')
my_jsons
```

```python
textfile = open(API_name+"_API_plantuml.txt", "w")
textfile.write('@startuml\n')
for key, value in object_dict.items():
    textfile.write('class '+str(key) +'{\n')
    for i in range (len(value)):
        textfile.write(value[i] + "\n")
    textfile.write('}\n')
#textfile.write('}\n')
textfile.write('@enduml')
textfile.close()
```

## Write it to a .csv file

 a .csv file with a ; seperator between key and values

```python
with open(API_name+"_data.csv", mode="w") as pv:
    for key, value in object_dict.items():
        pv.write(str(key)+";")
        for i in range (len(value)-1):
            pv.write(value[i]+",")#write values + a comma  
        pv.write(value[-1]+"\n") #write last value (and no comma) 
```
