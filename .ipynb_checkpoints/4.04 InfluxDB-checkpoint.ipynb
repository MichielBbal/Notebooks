### 2a. Getting data and parse it to a list

```python
# -*- coding: utf-8 -*-
"""Query data in InfluxDB."""

from influxdb import InfluxDBClient
import os

influxdb_host = "influxdb.sensemakersams.org"
influxdb_port = 443
influxdb_user = "openears"
influxdb_password = "0pen3ar5"
influxdb_dbname = "iot"

client = InfluxDBClient(
    influxdb_host,
    influxdb_port,
    influxdb_user,
    influxdb_password,
    influxdb_dbname,
    ssl=True,
    verify_ssl=True,
)

rs = client.query("SELECT * FROM /.*/ WHERE app_id='test'", epoch="ms")

# Get raw JSON response.
raw = rs.raw

# Get points.
# The measurement in InfluxDB can be specified. The format of the measurements is app_id/dev_id.
# points = rs.get_points(measurement='test/test', tags={'dev_id': 'test'})
points = rs.get_points()

for point in points:
    print(f"Time: {point['time']}, Temperature: {point['temperature']}")
```

```python
# -*- coding: utf-8 -*-
"""Query data in InfluxDB."""

from influxdb import InfluxDBClient
import os

influxdb_host = "influxdb.sensemakersams.org"
influxdb_port = 443
influxdb_user = "openears"
influxdb_password = "0pen3ar5"
influxdb_dbname = "iot"

client = InfluxDBClient(
    influxdb_host,
    influxdb_port,
    influxdb_user,
    influxdb_password,
    influxdb_dbname,
    ssl=True,
    verify_ssl=True,
)

rs = client.query("SELECT * FROM /.*/ WHERE app_id='openears'", epoch="ms")

# Get raw JSON response.
raw = rs.raw

# Get points.
# The measurement in InfluxDB can be specified. The format of the measurements is app_id/dev_id.
points = rs.get_points(measurement='openears/OE001', tags={'dev_id': 'OE001'})
#points = rs.get_points()

for point in points:
    print(f"Time: {point['time']}")
print(points)
```

### 2b. Writing to a pandas df -issue

```python
# -*- coding: utf-8 -*-
"""Read data from InfluxDB into Pandas DataFrame."""

from influxdb import DataFrameClient
import os

influxdb_host = "influxdb.sensemakersams.org"
influxdb_port = 443
influxdb_user = "openears"
influxdb_password = '0pen3ar5'
influxdb_dbname = "iot"
#client = InfluxDBClient(influxdb_host, influxdb_port, influxdb_user, influxdb_password, influxdb_dbname, ssl=True,verify_ssl=True, headers={'Accept': 'application/json'}, gzip=True)

client = DataFrameClient(
    influxdb_host,
    influxdb_port,
    influxdb_user,
    influxdb_password,
    influxdb_dbname,
    ssl=True,
    verify_ssl=True,
)

# Query returns a dictionary of dataframes.
rs = client.query(
    "SELECT * FROM /.*/ WHERE app_id='openears'", chunked=True, chunk_size=10000
)
#print(rs.keys())

# The measurement in InfluxDB can be specified. The format of the measurements is app_id/dev_id.
df = rs["openears/OE001"]
print(df.tail())
```
