## Basics: Mean, Median and Mode

```python
# Calculating the Mean
mean = (5+6+7)/3
print(mean)

#Calculating the Mean of a list
my_list = [5,6,7]
sum(my_list) # use the built in sum function to create the sum of a list and divide it by length
mean = sum(my_list)/len(my_list)
print(mean)
```

```python
# two ways of calculating the Median: 1) sort + 1/2 length and 2) 

my_list = [7,6,5]
my_list.sort()#(key=None en reverse=False) 
x = len(my_list)
print(my_list[1]) #sort the list and print the middle value of the list
#print(x)

import numpy
print(numpy.median(my_list)) #float
```

```python
# The mode is the value that appears most in the list. 
# We will use scipy to return the mode
from scipy import stats
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
x = stats.mode(speed)
print(x)
```

##  Numpy for Standard Deviation, Variance and Percentile

<b>Numpy</b> or Numerical Python is a widely used Python module for maths / statistics

Standard Deviation is a measure how spread out the values are.
Variance is another measure how spread out the values are. 

Standard Deviation = σ (sigma) or numpy.std
Variance = σ2 or numpy.var
Sqrt of Variance = Standard Deviation 

What is the 90. percentile? That means that 90% is lower than a certain value.

```python
speed = [86,87,88,86,87,85,86]
print(numpy.std(speed)) # print the standard deviation
print(numpy.var(speed)) # print the variance

# Percentile
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
print(numpy.percentile(ages, 75))
```

## Counter module

```python
# Learning Counter module - to be replaced to data or something

from collections import Counter

exampleList = [1,2,5,34,2,13,6,7,5,2,34,9,10,12,0,2,2]
counts = Counter(exampleList) # this creates a dictionary with key:value
print(counts)
#print(counts[34])#prints the value of key 34
print(list(counts.elements())) #prints a list of the elements in the list
print(counts.most_common(2)) # prints the 2 most common pairs
# you can also do z= counts+counts2 (first and second list)
```

## Matplotlib sub-plots

## Matplotlib animation

```python
%matplotlib notebook
#source = https://www.youtube.com/watch?v=Ercd-Ip5PfQ&t=920s - Corey Schafer
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []
y_vals2 = []

index = count()

def animate(i):
    x_vals.append(next(index))
    y_vals.append(random.randint(0,5))
    y_vals2.append(random.randint(0,5))
    plt.cla()
    plt.plot(x_vals, y_vals, label='Channel 1')
    plt.plot(x_vals, y_vals2, label='Channel2')
    
    plt.legend(loc='upper left')

ani = FuncAnimation(plt.gcf(), animate, interval=1000) #gcf get current function

#plt.tight_layout()
plt.show()
```

## Matplotlib animation #2

We will live plot the cpu usage. 

```python
#https://www.youtube.com/watch?v=GIywmJbGH-8&t=330s
%matplotlib notebook
import time
import psutil
import matplotlib.pyplot as plt

plt.rcParams['animation.html'] = 'jshtml'

fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()
i = 0
x, y = [],[]

while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    
    ax.plot(x,y, color='r')
    
    fig.canvas.draw()
    
    ax.set_xlim(left=max(0, i-50), right=i+30)
    
    time.sleep(0.2)
    i += 1
```

```python
#source: https://matplotlib.org/3.2.1/gallery/animation/simple_anim.html
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)
line, = ax.plot(x, np.sin(x))


def init():  # only required for blitting to give a clean slate.
    line.set_ydata([np.nan] * len(x))
    return line,


def animate(i):
    line.set_ydata(np.sin(x + i / 100))  # update the data.
    return line,


ani = animation.FuncAnimation(
    fig, animate, init_func=init, interval=2, blit=True, save_count=50)

# To save the animation, use e.g.
#
# ani.save("movie.mp4")
#
# or
#
# writer = animation.FFMpegWriter(
#     fps=15, metadata=dict(artist='Me'), bitrate=1800)
# ani.save("movie.mp4", writer=writer)

plt.show()
```

## Matplotlib animation: Die Rolls
#see Intro to Python p231-234

```python
%matplotlib notebook 
"""Dynamically graphing frequencies of die rolls"""
from matplotlib import animation
import matplotlib.pyplot as plt
import random
import seaborn as sns
import sys
from time import sleep

def update(frame_number, rolls, faces, frequencies):
    """configure bar plot contents for each animation frame"""
    #roll die and update frequencies
    for i in range(rolls):
        time.sleep(1)
        frequencies[random.randrange(1,7) - 1] +=1

        #reconfigure plot for update die frequencies
        plt.cla() #clear
        axes = sns.barplot(faces, frequencies, palette='bright') #new bars
        axes.set_title(f'Die frequencies for {sum(frequencies):,} rolls')
        axes.set(xlabel = 'Die Value', ylabel='Frequency')
        axes.set_ylim(top=max(frequencies)*1.1)
        
        #display frequenci & percentage abvoe each patch
        for bar, frequency in zip(axes.patches, frequencies):
            text_x = bar.get_x() + bar.get_width() / 2.0
            text_y = bar.get_height()
            text = f'{frequency:,}\n{frequency / sum(frequencies):.3%}' 
            axes.text(text_x, text_y, text, ha= 'center', va='bottom')
            
#read command-line args
number_of_frames = 10#int(sys.argv[1])
rolls_per_frame = 10#int(sys.argv[2])

sns.set_style('whitegrid')
figure = plt.figure("Rolling a six-sided die")
values = list(range(1,7))
frequencies = [0]*6

#configure
die_animation = animation.FuncAnimation(
    figure, update, repeat = False, frames=number_of_frames, interval=33,
    fargs=(rolls_per_frame, values, frequencies))

plt.show()
            
```
