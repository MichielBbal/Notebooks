## 2. Audio analysis with Librosa

Librosa is a python package for music and audio analysis. It provides the building blocks necessary to create audio information retrieval systems.

source: https://librosa.org/doc/latest/index.html

```python
wav_files
```

```python
# select an item in the list above:
audio_data = wav_files[10] #change the number between 0 and x to select one file
```

```python
from IPython import display
print(audio_data)
display.Audio(audio_data)
```

```python
#load the wav in librosa library
import librosa
x , sr = librosa.load(audio_data) #sr = sampling rate
#print the sample rate 
librosa.get_samplerate(audio_data)
```

```python
#creating a waveplot 
%matplotlib inline
import matplotlib.pyplot as plt
import librosa.display
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

```python
#creating a spectrogram
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```
