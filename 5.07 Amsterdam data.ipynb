# 5.07 Amsterdam data

Data taken from https://api.data.amsterdam.nl/api/

```python
import requests
import json
url = "https://api.data.amsterdam.nl/v1/gebieden/buurten/" #it's best to define the url seperately
response = requests.get(url)
todos = json.loads(response.text) #with json.loads we can convert the json object to a list object
type(todos)
todos
```
