## Methods

```python
import operator
from zeep import Client

methods_list = []
wsdl = "C:\\Users\\31653\\Desktop\\DeAlliantie\movinu_wos_outgoing_wsdl.xml"
client = Client(wsdl=wsdl)
for service in client.wsdl.services.values():
    print("service :", service.name)
    for port in service.ports.values():
        operations = sorted(
            port.binding._operations.values(),
            key=operator.attrgetter('name'))

        for operation in operations:
            #print(" method :", operation.name)
            methods_list.append(operation.name)
            #print("  input :", operation.input.signature())
            #print(" output :", operation.output.signature())
print(len(methods_list))
print(methods_list)
```

### Data objects and attributes

```python
# The new main script! TO DO: create one super script that gives me everything i need...
#source: https://stackoverflow.com/questions/50089400/introspecting-a-wsdl-with-python-zeep?noredirect=1&lq=1
from zeep import Client
from pprint import pprint

wsdl = "C:\\Users\\31653\\Desktop\\DeAlliantie\movinu_wos_outgoing_wsdl.xml"
client = Client(wsdl)


def parseElements(elements):
    all_elements = {}
    for name, element in elements:
        all_elements[name] = {}
        all_elements[name]['optional'] = element.is_optional
        if hasattr(element.type, 'elements'):
            all_elements[name]['type'] = parseElements(
                element.type.elements)
        else:
            all_elements[name]['type'] = str(element.type)

    return all_elements

interface = {}
for service in client.wsdl.services.values():
    interface[service.name] = {}
    for port in service.ports.values():
        interface[service.name][port.name] = {}
        operations = {}
        for operation in port.binding._operations.values():
            print(operation.name)
            operations[operation.name] = {}
            operations[operation.name]['input'] = {}
            elements = operation.input.body.type.elements
            operations[operation.name]['input'] = parseElements(elements)
            interface[service.name][port.name]['operations'] = operations
            print(elements)
            print('-------------')
            print(interface[service.name][port.name]['operations'])
```

```python
my_dict =interface[service.name][port.name]['operations']
list(my_dict)
```

```python
# The script to get the data out manually
my_methods =interface[service.name][port.name]['operations']
l = list(my_methods)
for i in range(len(l)):
    my_dict =interface[service.name][port.name]['operations'][l[i]]['input']
    for key, values in my_dict.items():
        print(key)
        dict2 = values
        dict2 = dict2['type']
        for key,values in dict2.items():
            print("    "+key)
            dict3 = values
            dict3 = dict3['type']
            try:
                for key,values in dict3.items():
                    print("        "+key)
                    dict4 = values
                    dict4 = dict4['type']
                try:
                    for key, values in dict4.items():
                        print("            "+key)
                        #print(values)
                except:
                    pass
            except:
                pass
```
