```python
#get the OpenAPI specification (.yaml file) from source
import wget

url = 'https://github.com/vereniging-corponet/vera-openapi/blob/main/Informatiedomeinen/Relaties.yaml' #change if necessary
filename = wget.download(url)
```

```python
# show all .yaml files in current directory
import glob
my_yamls=glob.glob('*.yaml')
my_yamls
```

```python
# import the yaml file as a python dictionary
import yaml

# load the spec file and read the yaml to create a python dictionary 'spec'
with open('Relaties.yaml') as f:
    spec = yaml.safe_load(f.read())
```

```python
# show specification
list(spec)
```

```python
# print OAS-version, API name and API version
OAS_version =spec['openapi']
API_name = spec['info']['title']
API_name = API_name.replace(" ", "_")
API_version =spec['info']['version']
print(OAS_version, API_name, API_version)
```

## 2. Getting the paths and endpoints

```python
#inspecting the paths
spec['paths']
```

```python
#getting the paths
paths = list(spec['paths'])
len(paths)
print(paths)
```

```python
#getting the paths and http verbs
endpoints_dict = {}
for i in range (len(paths)):
    endpoints_dict[paths[i]]=list(spec['paths'][paths[i]])
endpoints_dict 
```

```python
#printing the paths, verbs and descriptions 
paths = list(spec['paths'])
for i in range (len(paths)):
    verbs = list(spec['paths'][paths[i]])
    for j in range (len(verbs)):
        description_string = spec['paths'][paths[i]][verbs[j]]['description']
        description_string = description_string.strip('\n')
        print(paths[i], verbs[j], description_string)
```

```python
#printing the paths, verbs and descriptions in a pretty table
from prettytable import PrettyTable
    
my_table = PrettyTable()

my_table.field_names = ["Path", "Verb", "Description"]
endpoints = list(spec['paths'])
for i in range (len(endpoints)):
    verbs = list(spec['paths'][endpoints[i]])
    for j in range (len(verbs)):
        description_string = spec['paths'][endpoints[i]][verbs[j]]['description']
        description_string = description_string.strip('\n')
        #print(verbs[j], endpoints[i], description_string)
        my_table.add_row([endpoints[i], verbs[j], description_string[:80]])
print(my_table)
```

## 3. Getting the data objects & attributes

```python
#getting the objects
objects = list(spec['components']['schemas'])
print(len(objects))
objects.sort()
print(objects)
```

```python
# lambda function to remove '-sleutels' from the list of objects
objects = [x for x in objects if "-sleutels" not in x ]
print(objects)
```

```python
#creating a dictionary with objects and their attributes
object_dict ={}
#objects = list(spec['components']['schemas'])
for i in range (len(objects)):
    try:
        object_dict[objects[i]]=list(spec['components']['schemas'][objects[i]]['properties'])#['properties']
    except:
        my_list= []
        my_dict = spec['components']['schemas'][objects[i]]['allOf'][0]
        for key, value in my_dict.items():
            my_string = str(value)
        my_string = my_string.replace('#/components/schemas/', 'allOf: ')
        my_list.append(my_string)
        object_dict[objects[i]] = my_list+list(spec['components']['schemas'][objects[i]]['allOf'][1]['properties'])
object_dict
```

## 4. Saving as plantuml, .csv and .json

### 4a. Creating a plantuml file

```python
#create a plantuml file
textfile = open(API_name+"_API_plantuml.txt", "w")
textfile.write('@startuml\n')
for key, value in object_dict.items():
    textfile.write('class '+str(key) +'{\n')
    for i in range (len(value)):
        textfile.write(value[i] + "\n")
    textfile.write('}\n')
#textfile.write('}\n')
textfile.write('@enduml')
textfile.close() 
```

### 4b. Saving as a .csv file

```python
with open(API_name+"_API_objecten.csv", mode="w") as my_csv:
    for key, value in object_dict.items():
        my_csv.write(str(key)+";")
        for i in range (len(value)-1):
            my_csv.write(value[i]+",")#write values + a comma  
        my_csv.write(value[-1]+"\n") #write last value (and no comma)
```
