```python
import requests
def weersverwachting(api_key=None, plaats=None):
    assert api_key is not None, 'Geef je API key op'
    plaats = plaats or 'Amsterdam'
    
    BASE_URL = 'https://weerlive.nl/api/json-data-10min.php?'
    verwachting = requests.get(f"{BASE_URL}key={api_key}&locatie={plaats}").json().get('liveweer')
    
    return verwachting #[0]
weersverwachting('demo' )
#def samenvatting(api_key=None, plaats=None):
```

delete(url, args)	Sends a DELETE request to the specified url <br>

            head(url, args)	Sends a HEAD request to the specified url<br>
    patch(url, data, args)	Sends a PATCH request to the specified url<br>
post(url, data, json, args)	Sends a POST request to the specified url <br>
    put(url, data, args)	Sends a PUT request to the specified url<br>
request(method, url, args)	Sends a request of the specified method to the specified url

### Reqbin
source: https://reqbin.com/req/python/5k564bhv/get-request-bearer-token-authorization-header-example

```python
import requests
from requests.structures import CaseInsensitiveDict

url = "https://reqbin.com/echo/get/json"

headers = CaseInsensitiveDict()
headers["Accept"] = "application/json"
headers["Authorization"] = "Bearer {token}"


resp = requests.get(url, headers=headers)

print(resp.status_code)
```

```python
print(dir(resp))
```

```python
print(resp.headers)
print('-------------------------')
print(resp.content)
```
