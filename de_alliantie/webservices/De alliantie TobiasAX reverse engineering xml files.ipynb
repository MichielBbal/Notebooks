## Onderhoudsverzoeken

```python
import xml.etree.ElementTree as ET #import the library 
tree = ET.parse('CreateServiceTask.xml') #parse the XML
ET.dump(tree) #print the xml file
```

```python
my_list =[]
for elem in tree.iter():
    my_list.append(elem.tag)
my_clean_list = list(dict.fromkeys(my_list))
my_clean_list
```

```python
my_clean_list = [x for x in my_clean_list if 'http://schemas.datacontract.org' in x]
my_clean_list
```

```python
attribute_list =[]
for i in range(len(my_clean_list)):
    my_string = my_clean_list[i]
    l = my_string.split("}")
    attribute = l[1]
    attribute_list.append(attribute)
attribute_list
```

```python
len(attribute_list)
```

```python
with open("TobiasAX_CreateServiceTask.csv", mode="w") as csv_file:
    for i in range(len(attribute_list)):
        csv_file.write(attribute_list[i]+"\n")
```
